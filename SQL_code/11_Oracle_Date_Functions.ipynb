```python
SELECT
EXTRACT(year FROM order_date) year,
EXTRACT (month FROM order_date) month,
COUNT (order_id) order_count
FROM orders
GROUP BY
EXTRACT(YEAR FROM order_date),
EXTRACT(MONTH FROM order_date)
ORDER BY year DESC, month;
```
